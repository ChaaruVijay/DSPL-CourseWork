```python
import pandas as pd
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train=pd.read_csv('train_cleaned.csv')
train
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='cluster_category', y='fresh_sales', data=train)
plt.title('Distribution of Fresh Sales by Customer Segment')
plt.xlabel('Cluster Category')
plt.ylabel('Fresh Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='cluster_category', y='luxury_sales', data=train)
plt.title('Distribution of luxury Sales by Customer Segment')
plt.xlabel('Cluster Category')
plt.ylabel('luxury Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='cluster_category', y='dry_sales', data=train)
plt.title('Distribution of Dry Sales by Customer Segment')
plt.xlabel('Cluster Category')
plt.ylabel('Dry Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df_melted = train.melt(id_vars='cluster_category', value_vars=['luxury_sales', 'fresh_sales', 'dry_sales'], var_name='sales_type', value_name='sales_amount')

# Create the grouped boxplot
plt.figure(figsize=(12, 8))
sns.barplot(x='cluster_category', y='sales_amount', hue='sales_type', data=df_melted)

plt.title('Distribution of Sales by Cluster and Product Type')
plt.xlabel('Cluster Category')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sales Type')
plt.tight_layout()
plt.show()
```
